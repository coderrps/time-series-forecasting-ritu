```python
# Exponential Smoothing Models
def exponential_smoothing(train, test, trend='add', seasonal='add', seasonal_periods=12):
    model = ExponentialSmoothing(train, trend=trend, seasonal=seasonal, seasonal_periods=seasonal_periods)
    fit_model = model.fit()
    predictions = fit_model.forecast(len(test))
    return predictions

# Regression Model
def linear_regression(train, test):
    X_train, y_train = np.arange(len(train)).reshape(-1, 1), train.values
    X_test = np.arange(len(train), len(train) + len(test)).reshape(-1, 1)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    return predictions

# Naïve Forecast Model
def naive_forecast(train, test):
    last_observation = train.iloc[-1]
    predictions = [last_observation] * len(test)
    return predictions

# Simple Average Model
def simple_average(train, test):
    average_value = train.mean()
    predictions = [average_value] * len(test)
    return predictions

# Evaluate models using RMSE
def evaluate_model(predictions, actual):
    rmse = sqrt(mean_squared_error(actual, predictions))
    return rmse

# Build and evaluate Exponential Smoothing models
predictions_exp_smoothing = exponential_smoothing(train['Sparkling'], test['Sparkling'])
rmse_exp_smoothing = evaluate_model(predictions_exp_smoothing, test['Sparkling'])
print(f'Exponential Smoothing RMSE: {rmse_exp_smoothing}')

# Build and evaluate Linear Regression model
predictions_regression = linear_regression(train['Sparkling'], test['Sparkling'])
rmse_regression = evaluate_model(predictions_regression, test['Sparkling'])
print(f'Linear Regression RMSE: {rmse_regression}')

# Build and evaluate Naïve Forecast model
predictions_naive = naive_forecast(train['Sparkling'], test['Sparkling'])
rmse_naive = evaluate_model(predictions_naive, test['Sparkling'])
print(f'Naïve Forecast RMSE: {rmse_naive}')

# Build and evaluate Simple Average model
predictions_avg = simple_average(train['Sparkling'], test['Sparkling'])
rmse_avg = evaluate_model(predictions_avg, test['Sparkling'])
print(f'Simple Average RMSE: {rmse_avg}')
```

```python
def check_stationarity(timeseries):
    # Perform ADF test
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])

    # Visualize rolling statistics
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()

    plt.figure(figsize=(12, 6))
    plt.plot(timeseries, label='Original')
    plt.plot(rolmean, label='Rolling Mean')
    plt.plot(rolstd, label='Rolling Std')
    plt.legend()
    plt.title('Rolling Mean and Standard Deviation')
    plt.show()

# Check stationarity for 'Sparkling' column
check_stationarity(merged_df['Sparkling'])

# If data is non-stationary, take appropriate steps to make it stationary
# Example: Differencing
merged_df['Sparkling_diff'] = merged_df['Sparkling'] - merged_df['Sparkling'].shift(1)
merged_df.dropna(inplace=True)

# Check stationarity for differenced 'Sparkling' column
check_stationarity(merged_df['Sparkling_diff'])
```

```python
def check_stationarity(timeseries, alpha=0.05):
    # Perform ADF test
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])

    # Compare p-value with alpha
    print(f'Significance Level (alpha): {alpha}')
    if result[1] <= alpha:
        print('Reject the null hypothesis - Data is stationary.')
    else:
        print('Fail to reject the null hypothesis - Data is non-stationary.')

    # Visualize rolling statistics
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()

    plt.figure(figsize=(12, 6))
    plt.plot(timeseries, label='Original')
    plt.plot(rolmean, label='Rolling Mean')
    plt.plot(rolstd, label='Rolling Std')
    plt.legend()
    plt.title('Rolling Mean and Standard Deviation')
    plt.show()

# Check stationarity for 'Sparkling' column with alpha=0.05
check_stationarity(merged_df['Sparkling'], alpha=0.05)
```
